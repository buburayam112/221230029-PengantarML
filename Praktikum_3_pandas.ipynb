```python
'''Analisis Dataset Retail (Versi Modifikasi)'''
# Import libraries
import numpy as np
import pandas as pd

# Membuat dataset simulasi transaksi retail
np.random.seed(123)  # ubah seed agar hasil acak berbeda
tanggal = pd.date_range(start='2023-01-05', end='2023-04-05', freq='D')

data_retail = pd.DataFrame({
    'tanggal': np.random.choice(tanggal, 950),  # ubah jumlah data sedikit
    'produk_id': np.random.randint(1, 12, 950),  # 11 produk
    'jumlah': np.random.randint(1, 12, 950),     # jumlah pembelian 1–11
    'harga': np.random.uniform(15, 120, 950),    # harga 15–120
    'pelanggan_id': np.random.randint(1, 121, 950)  # 120 pelanggan
})

# Tambahkan kolom total pendapatan
data_retail['pendapatan'] = data_retail['jumlah'] * data_retail['harga']

# === SOLUSI 1: Hitung total pendapatan per produk ===
pendapatan_per_produk = (
    data_retail.groupby('produk_id')['pendapatan']
    .sum()
    .sort_values(ascending=False)
)

#  Temukan 5 pelanggan dengan pembelian tertinggi ===
pelanggan_terbaik = (
    data_retail.groupby('pelanggan_id')['pendapatan']
    .sum()
    .nlargest(5)
)

#  Analisis pendapatan harian ===
pendapatan_harian = (
    data_retail.groupby('tanggal')['pendapatan']
    .sum()
    .sort_index()
)

#  Deteksi anomali pada kolom jumlah pembelian ===
def deteksi_anomali_jumlah(df, batas_z=2.5):
    rata2 = df['jumlah'].mean()
    std = df['jumlah'].std()
    df['zscore_jumlah'] = (df['jumlah'] - rata2) / std
    anomali = df[np.abs(df['zscore_jumlah']) > batas_z]
    return anomali

data_anomali = deteksi_anomali_jumlah(data_retail)

print("=== HASIL ANALISIS DATA RETAIL ===")
print("\n1️⃣ Total Pendapatan per Produk:")
print(pendapatan_per_produk)

print("\n2️⃣ Lima Pelanggan dengan Pembelian Tertinggi:")
print(pelanggan_terbaik)

print("\n3️⃣ Pendapatan Harian (10 data pertama):")
print(pendapatan_harian.head(10))

print(f"\n4️⃣ Jumlah Anomali Terdeteksi: {len(data_anomali)}")
print("Contoh Data Anomali:")
print(data_anomali[['tanggal', 'produk_id', 'jumlah', 'zscore_jumlah']].head())

# === BONUS ANALISIS ===
print("\n=== ANALISIS TAMBAHAN ===")

# Tren pendapatan bulanan
data_retail['bulan'] = data_retail['tanggal'].dt.month
pendapatan_bulanan = (
    data_retail.groupby('bulan')['pendapatan']
    .sum()
    .round(2)
)
print("\nPendapatan Bulanan:")
print(pendapatan_bulanan)

# Segmentasi pelanggan berdasarkan perilaku pembelian
statistik_pelanggan = (
    data_retail.groupby('pelanggan_id')
    .agg({
        'pendapatan': ['sum', 'mean', 'count'],
        'jumlah': 'mean'
    })
    .round(2)
)
print("\nStatistik Pelanggan:")
print(statistik_pelanggan.head())

# === PENGUJIAN (Assertions) ===
assert len(pendapatan_per_produk) <= 11, "Jumlah produk tidak boleh lebih dari 11"
assert len(pelanggan_terbaik) == 5, "Harus ada 5 pelanggan terbaik"
assert len(pendapatan_harian) > 0, "Pendapatan harian tidak boleh kosong"

print("\n✅ Semua pengujian berhasil! Analisis berjalan dengan benar.")
```
