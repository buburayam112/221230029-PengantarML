```python
'''Manipulasi Data List untuk Preprocessing'''
# Data mentah dari sensor (dalam derajat Celcius)
raw_data = [19.8, 21.5, 22.7, 23.9, 24.4, 25.8, 26.3, 22.1, 20.9, 23.4]

# Filter data yang berada di range 21–24 derajat
filtered_data = [x for x in raw_data if 21 <= x <= 24]

# Normalisasi data ke range 0–1 menggunakan min-max scaling
min_val = min(raw_data)
max_val = max(raw_data)
normalized_data = [(x - min_val) / (max_val - min_val) for x in raw_data]

# Hitung moving average dengan window size 4
def moving_average(data, window_size):
    moving_avgs = []
    for i in range(len(data) - window_size + 1):
        window = data[i:i + window_size]
        avg = sum(window) / window_size
        moving_avgs.append(avg)
    return moving_avgs

ma_result = moving_average(raw_data, 4)

print("Raw data:", raw_data)
print("Filtered data (21–24):", filtered_data)
print("Normalized data:", [round(x, 3) for x in normalized_data])
print("Moving average (window=4):", [round(x, 3) for x in ma_result])
```

### OPERASI DICTIONARY ###

```python
'''SOLUSI: Processing Dataset untuk Machine Learning'''
# Dataset contoh untuk klasifikasi sederhana
dataset = [
    {'features': [2.5, 4.1, 3.3, 1.8], 'label': 'type_X'},
    {'features': [3.8, 2.6, 4.7, 3.1], 'label': 'type_Y'},
    {'features': [2.9, 3.7, 3.0, 2.2], 'label': 'type_X'},
    {'features': [4.4, 3.3, 5.2, 3.6], 'label': 'type_Y'}
]

# Tahap 1: Kelompokkan data berdasarkan label
def group_by_label(data):
    grouped = {}
    for item in data:
        label = item['label']
        if label not in grouped:
            grouped[label] = []
        grouped[label].append(item['features'])
    return grouped

grouped_data = group_by_label(dataset)

# Tahap 2: Hitung rata-rata tiap fitur untuk setiap kelas
def average_features_per_class(grouped_data):
    averages = {}
    for label, features_list in grouped_data.items():
        n_features = len(features_list[0])
        sums = [0] * n_features
        for features in features_list:
            for i, val in enumerate(features):
                sums[i] += val
        averages[label] = [s / len(features_list) for s in sums]
    return averages

avg_features = average_features_per_class(grouped_data)

# Tampilkan hasil pengelompokan dan rata-rata
print("\nGrouped data by label:")
for label, features in grouped_data.items():
    print(f"{label}: {features}")

print("\nAverage features per class:")
for label, avg in avg_features.items():
    print(f"{label}: {[round(x, 3) for x in avg]}")

# TEST ASSERTIONS (validasi sederhana)
assert 'type_X' in grouped_data, "Dataset harus berisi type_X"
assert 'type_Y' in grouped_data, "Dataset harus berisi type_Y"
assert all(isinstance(f, float) for item in dataset for f in item['features']), "Semua fitur harus berupa angka float"

print("\n✅ Semua assertions berhasil! Solusi benar.")
```
