```python
import matplotlib.pyplot as plt
import numpy as np
```

1. Basic Plot Setup

```python
print("=== BASIC SETUP DAN LINE PLOT ===")

# Data contoh
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='sin(x)', color='blue', linewidth=2, linestyle='-')
plt.xlabel('X Axis Label', fontsize=12)
plt.ylabel('Y Axis Label', fontsize=12)
plt.title('Basic Line Plot dengan Matplotlib', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

2. Multiple Plot Dalam Satu Figure

```python
print("=== MULTIPLE PLOTS DALAM SATU FIGURE ===")

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x/2)

plt.figure(figsize=(12, 8))

# Plot 1
plt.subplot(2, 2, 1)  # (rows, columns, position)
plt.plot(x, y1, 'r-', label='sin(x)')
plt.title('Sine Wave')
plt.legend()

# Plot 2
plt.subplot(2, 2, 2)
plt.plot(x, y2, 'g--', label='cos(x)')
plt.title('Cosine Wave')
plt.legend()

# Plot 3
plt.subplot(2, 2, 3)
plt.plot(x, y3, 'b:', label='tan(x/2)')
plt.title('Tangent Wave')
plt.legend()

# Plot 4
plt.subplot(2, 2, 4)
plt.plot(x, y1, 'r-', label='sin(x)')
plt.plot(x, y2, 'g--', label='cos(x)')
plt.plot(x, y3, 'b:', label='tan(x/2)')
plt.title('Combined Plot')
plt.legend()

plt.tight_layout()
plt.show()
```

3. Scatter Plot

```python
print("=== SCATTER PLOT ===")

np.random.seed(42)
x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)
colors = np.random.rand(100)
sizes = 100 * np.random.rand(100)

plt.figure(figsize=(10, 6))
scatter = plt.scatter(x, y, c=colors, s=sizes, alpha=0.7, cmap='viridis')
plt.colorbar(scatter, label='Color Intensity')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Scatter Plot dengan Color Mapping')
plt.grid(True, alpha=0.3)
plt.show()
```

4. Bar Plot dan Histogram

```python
print("=== BAR PLOT DAN HISTOGRAM ===")

# Data untuk bar plot
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 45, 56, 12, 67]

# Data untuk histogram
data_hist = np.random.normal(0, 1, 1000)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Bar plot
ax1.bar(categories, values, color=['red', 'blue', 'green', 'orange', 'purple'])
ax1.set_title('Bar Plot Example')
ax1.set_xlabel('Categories')
ax1.set_ylabel('Values')

# Histogram
ax2.hist(data_hist, bins=30, alpha=0.7, color='skyblue', edgecolor='black')
ax2.set_title('Histogram Example')
ax2.set_xlabel('Value')
ax2.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

5. Customization Advanced

```python
print("=== ADVANCED CUSTOMIZATION ===")

x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(12, 6))

# Plot dengan berbagai customizations
plt.plot(x, y1, label='sin(x)', color='red', linewidth=2,
         marker='o', markersize=4, markevery=5, alpha=0.7)
plt.plot(x, y2, label='cos(x)', color='blue', linewidth=2,
         linestyle='--', alpha=0.7)

# Customize axes
plt.xlim(0, 2*np.pi)
plt.ylim(-1.5, 1.5)
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
           ['0', '/2', '', '3/2', '2'])
plt.yticks([-1, -0.5, 0, 0.5, 1])

# Add grid and legend
plt.grid(True, alpha=0.3, linestyle=':')
plt.legend(loc='upper right', fontsize=12, frameon=True, fancybox=True, shadow=True)

# Add text annotation
plt.annotate('Maximum sin(x)', xy=(np.pi/2, 1), xytext=(np.pi/2 + 0.5, 0.8),
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.2'),
             fontsize=10)

plt.title('Advanced Plot Customization', fontsize=14, pad=20)
plt.xlabel('X Axis (Radians)', fontsize=12)
plt.ylabel('Y Axis (Amplitude)', fontsize=12)

plt.show()
```
